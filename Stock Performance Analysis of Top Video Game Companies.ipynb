```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# read csv file data
data = pd.read_csv(r"C:\Users\Manikanta\Downloads\VideoGameStocks.csv")
data
```

```python
# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Check for missing or invalid data
data.dropna(inplace=True)  # Drop rows with missing values
data = data.drop_duplicates()  # Remove duplicate rows
```

```python
# extract year and month from date column
data['Year']=data['Date'].dt.year
data['Month']=data['Date'].dt.month_name()
data
```

```python
# Add a column for Monthly Returns if the 'Close' column exists
data['Monthly Return (%)'] = data['Close'].pct_change() * 100
data
```

```python
# average closing price for each company and round it into two decimal points
avg_close_price_by_company = round(data.groupby('Company')['Close'].mean(),2)
avg_close_price_by_company
```

```python
# top 5 companies by trading volume
top_5_volume = data.groupby('Company')['Volume'].sum().sort_values(ascending=False).head(5)
top_5_volume
```

```python
# calculate overall mean and median closing price
mean_close = round(np.mean(data['Close']),2)
median_close = round(np.median(data['Close']),2)
print(f'mean_close = {mean_close}\nmedian_close = {median_close}')
```

```python
#Line Plot of Average Closing Prices by Year
avg_close_by_year = data.groupby('Year')['Close'].mean()
plt.figure(figsize=(8, 6))
avg_close_by_year.plot(kind='line', marker='o', color='blue')
plt.title('Average Closing Prices by Year')
plt.xlabel('Year')
plt.ylabel('Average Closing Price')
plt.show()
```

```python
# Bar Chart of Top 5 Companies by Trading Volume
plt.figure(figsize=(8, 6))
top_5_volume.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Top 5 Companies by Total Trading Volume')
plt.xlabel('Company')
plt.ylabel('Total Volume')
plt.show()
```
